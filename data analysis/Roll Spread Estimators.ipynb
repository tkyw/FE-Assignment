```python
def get_gamma(high0, low0, high1, low1):
    return (np.log(np.max([high0, high1]) / np.min([low0, low1]))) ** 2

def get_beta(high0, low0, high1, low1):
    return (np.log(high0 / low0) + np.log(high1 / low1)) ** 2

def get_alpha(beta, gamma):
    return ((np.sqrt(2 * beta)) - (np.sqrt(beta))) / (3 - (2 * np.sqrt(2))) - np.sqrt(gamma / (3 - (2 * np.sqrt(2))))

def spread(high0, high1, low0, low1):
    gamma = get_gamma(high0, low0, high1, low1)
    beta = get_beta(high0, low0, high1, low1)
    alpha = get_alpha(beta, gamma)
    return (2 * (np.exp(alpha) - 1)) / (1 + np.exp(alpha))
```
