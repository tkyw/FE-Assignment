```python
import pandas as pd 
import numpy as np
import os.path
```

```python
sheet_name=["Gold Futures Historical Data", "XAU USD Gold Spot US Dollar", 'US Dollar Index Futures' , "S&P index", "USD CNY", "FDR", "Crude Oil Price"]
```

```python

df = pd.read_excel(r"C:\Users\USER\Python Programme\uni\fe_assignment\Gold Futures Historical Data.xlsx", sheet_name=sheet_name, index_col=0, parse_dates=True)
df_future, df_spot, df_dollar, df_market_index, df_exchange, df_fdr, df_oil = df['Gold Futures Historical Data'], df['XAU USD Gold Spot US Dollar'], df['US Dollar Index Futures'], df['S&P index'], df['USD CNY'], df['FDR'], df['Crude Oil Price']
```

```python
dfs = [df_future, df_spot, df_dollar, df_market_index, df_exchange, df_fdr, df_oil]
keys = ["future", "spot", "dollar", "market", "exchange" ,"fdr", "oil"]
appended_keys = [""]
df_merge = pd.DataFrame()
for df, key in zip(dfs, keys):
    appended_keys.append(key)
    df_merge = df_merge.merge(df, left_index=True, right_index=True, how="outer")
df_merge
```

```python
nw_df_merge = (df_merge 
                .dropna(axis=0)
)
index= nw_df_merge.index
index
```

```python
def record_excel(df, sheet_name):
    df = df.loc[index, :]
    df.to_excel(wf, sheet_name=sheet_name)
```

```python
filename = "combine_data.xlsx"
for temp_df, name in zip(dfs, sheet_name):
    if not os.path.exists(filename):
        with pd.ExcelWriter(filename, mode='w', engine='openpyxl') as wf:
            record_excel(temp_df, name)
    else:
        with pd.ExcelWriter(filename, mode='a', engine='openpyxl') as wf:
            record_excel(temp_df, name)
```
