```python
import pandas as pd 
import numpy as np 
from data import get_data
```

```python
def unit_converter(value):
    units = {
        "K": 10 ** 3
    }
    return [float(v[:-1]) * units[v[-1]] for v in value]
```

```python
df = get_data("trading volume", "", [""])
```

```python
vol = df.apply(lambda x: unit_converter(x))
```

```python
future_spot = pd.read_excel(r"C:\Users\USER\Python Programme\uni\fe_assignment\spot_future.xlsx", sheet_name="deviation and cost", index_col= 0, parse_dates=True).iloc[:, :7]
future_spot
```

```python
vol_dollar = vol.sort_index(ascending=True).mul(future_spot['Spot'], axis=0)
vol_dollar
```

```python
future_spot.merge(vol_dollar, left_index=True, right_index=True).dropna(axis=0).to_clipboard(excel=True)
```
