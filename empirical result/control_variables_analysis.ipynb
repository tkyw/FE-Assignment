```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
plt.style.use("seaborn")
```

```python
df = pd.read_excel("C:\\Users\\USER\\Python Programme\\uni\\fe_assignment\\combine_data.xlsx", sheet_name = 0, index_col = 0, parse_dates = True)
df
```

```python
x, y = 2, 3
i = 1
fig, ax = plt.subplots(x, y, figsize=(15, 10))
values = ["Index", "Index", "Chinese Yen / US Dollar", "%", "%"]
for row in range(x):
    for col in range(y):
        if i != len(df.columns)-1:
            ax[row, col].plot(df[df.columns[i+1]])
            ax[row, col].set( ** {
                "title": df.columns[i+1].title(),
                "xlabel": "Year",
                "ylabel": values[i-1]
            })
            i += 1
        else:
            fig.delaxes(ax[row, col])
plt.tight_layout()
```
